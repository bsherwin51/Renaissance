```python
import numpy as np
import h5py as h5
import matplotlib.pyplot as plt
```

```python
data = {}
with h5.File('RP/catalog.h5', 'r') as fp:
    for g in fp:
        if data == {}:
            for field in fp[g]:
                data[field] = []
        for field in fp[g]:
            data[field] += fp[g][field][:].tolist()
for k,v in data.items():
    data[k] = np.array(data[k]).ravel()
print(data.keys())
```

```python
plt.scatter(data['mass'], data['Mstar'], alpha=0.1)
plt.xscale('log')
plt.yscale('log')
plt.xlim(1e7, 3e9)
plt.show()
```

```python
plt.scatter(data['mass'], data['SFR'], alpha=0.1)
plt.xscale('log')
plt.yscale('log')
plt.xlim(1e7, 3e9)
plt.show()
```

```python
plt.scatter(data['mass'], data['SFR']/data['Mstar'], alpha=0.1)
plt.xscale('log')
plt.yscale('log')
plt.xlim(1e7, 3e9)
plt.show()
```

```python
plt.scatter(data['Mstar'], data['SFR']/data['Mstar'], alpha=0.1)
plt.xscale('log')
plt.yscale('log')
plt.xlim(1e3, 5e7)
plt.show()
```

```python
plt.scatter(data['Mstar'], data['Mstar_pop2_young'], alpha=0.1)
plt.xscale('log')
plt.yscale('log')
plt.xlim(1e3, 5e7)
plt.show()
```
